### 1- Write a program to implement the example on p. 48 session 1 

```python
def fact(num):
    factorial = 1
    for i in range(1, num + 1):
        factorial = factorial * i
    return factorial

perm = fact(10) / (fact(4) * fact(3) * fact(2) * fact(1))
print('#permutations = ', perm)
```

### 2 - Assign different prices to 5 kinds of chocolate -> Type 1: 1 L.E., Type 2: 1.5, Type 3: 2.25, Type 4: 3, Type 5: 5 L.E. 

```python
import numpy as np
prices = np.array([1, 1.5, 2.25, 3, 5], dtype=np.float64)
#print('The price of each type is = ', prices)
for i in range(0, len(prices)):
    print('The price of type {} is = {}'.format(i+1, prices[i]))
```

### 3-The program creates combinations for boxes of 30 assorted pieces from the 5 kinds, print the content (no. from each kind) with the corresponding total price 


```python
from itertools import combinations_with_replacement
 
comb = combinations_with_replacement([1, 2, 3, 4, 5], 30)
 
count = 0
type1 = 0
type2 = 0
type3 = 0
type4 = 0
type5 = 0
priceOfBox = []

print('****************************************************************************************************')
print('                                 Box                                                           Price')
print('****************************************************************************************************')
for i in comb:
    for j in i:
        if j == 1:
            type1 = type1 + 1
        elif j == 2:
            type2 = type2 + 1
        elif j == 3:
            type3 = type3 + 1
        elif j == 4:
            type4 = type4 + 1
        elif j == 5:
            type5 = type5 + 1
    if type1 == 0 or type2 == 0 or type3 == 0 or type4 == 0 or type5 == 0:
        del(i) 
    else:
        totalP = type1 * prices[0] + type2 * prices[1] + type3 * prices[2] + type4 * prices[3] + type5 * prices[4]
        priceOfBox.append(totalP)
        count = count + 1
        print(i, ' * ', totalP)
        #print(' 1-> ', type1, ' 2-> ', type2, ' 3-> ', type3, ' 4-> ', type4, ' 5->', type5)
    
    type1 = 0
    type2 = 0
    type3 = 0
    type4 = 0
    type5 = 0
```

```python
print('The number of combination of boxes = ', count)
#print('combination type1 = {} and its price = {}'.format(comb, price1))
```

### 4- Compute the mean, variance and std. deviation of the price variable (hint: use numpy methods) 

```python
print('The mean of total prices in the box = ', np.mean(priceOfBox))

print('The variance of total prices in the box = ', np.var(priceOfBox))

print('The standard deviation of total prices in the box = ', np.std(priceOfBox))
```

### 5- Compute the probability of creating a box with equal number of pieces of each type

```python
countEqual = 0
for i in comb:
    for j in i:
        if j == 1:
            type1 = type1 + 1
        elif j == 2:
            type2 = type2 + 1
        elif j == 3:
            type3 = type3 + 1
        elif j == 4:
            type4 = type4 + 1
        elif j == 5:
            type5 = type5 + 1
    if type1 == type2 and type2 == type3 and type3 == type4 and type4 == type5:
        countEqual = countEqual + 1
    type1 = 0
    type2 = 0
    type3 = 0
    type4 = 0
    type5 = 0
print('Number of boxes with equal number of pieces of each type is = ', countEqual)
print('Probability of creating a box with equal number of pieces of each type is = ', countEqual / count)
```

### 6- Compute the probability of creating boxes with prices >= 100 L.E.

```python
countPrices = 0
for i in priceOfBox:
    if i >= 100:
        countPrices = countPrices + 1
print('Number of boxes with prices >= 100 L.E is = ', countPrices)
print('Probability of creating boxes with prices >= 100 L.E is = ', countPrices / count)
```
